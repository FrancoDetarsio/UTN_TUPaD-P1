1) Crear una lista con las notas de 10 estudiantes.
• Mostrar la lista completa.
• Calcular y mostrar el promedio.
• Indicar la nota más alta y la más baja.

```python
# Ejercicio 1

lista_notas = [8, 10, 9, 6.5, 7, 7.5, 9, 8, 5, 6]

# Inicializo variables para un correcto funcionamiento
promedio = 0
nota_alta = float('-inf')
nota_baja = float('inf')

# Calculo la cantidad de elementos de la lista para el bucle
cantidad_notas = len(lista_notas)


for alumno in range(cantidad_notas):
    # muestro el alumno y que nota le pertenece a este
    print(f"La nota del alumno {alumno + 1} es: {lista_notas[alumno]}")

    # voy sumando las notas para el promedio
    promedio += lista_notas[alumno]

    #voy verificando la nota mayor y la menor
    if lista_notas[alumno] > nota_alta:
        nota_alta = lista_notas[alumno]

    if lista_notas[alumno] < nota_baja:
        nota_baja = lista_notas[alumno]

print(f"El promedio es de: {promedio / cantidad_notas}") # Calculo el promedio dividiendo todo lo que sume por la cantidad de alumnos
print(f"La nota mas alta fue: {nota_alta}")
print(f"La nota mas baja fue: {nota_baja}.")
```

2) Pedir al usuario que cargue 5 productos en una lista.
• Mostrar la lista ordenada alfabéticamente. Investigue el uso del método sorted().
• Preguntar al usuario qué producto desea eliminar y actualizar la lista.

```python
# Ejercicio 2

lista_productos = []
cont = 0

# le pido al usuario uno por uno los articullos, y los voy sumando a la lista
for num_prod in range(1, 6):
    producto = input(f"Ingrese el {num_prod}º Producto: ")

    lista_productos += [producto]

# comando para acomodar alfabeticamente la lista
lista_ordenada = sorted(lista_productos)

# mustro la lista detallada y ordenada
print("Los articulos ingresados son:\n=============================")
for producto in lista_ordenada:
    cont += 1
    print(f"{cont}º {producto}")

# doy posibilidad de eliminar un articulo
print("Desea eliminar un articulo?")
eliminar = int(input("Ingrese un numero de articulo si desea borar alguno (1-5), presione 0 para salir"))

if (eliminar > 0) and (eliminar <= 5):
    del lista_ordenada[eliminar -1] # elimino el articulo que desea el usuario, restando 1 para que concuerde
    cont = 0
    print("Nueva lista:\n=============================")
    # imprimo nueva lista actualizada
    for producto in lista_ordenada:
        cont += 1
        print(f"{cont}º {producto}")

elif eliminar == 0:
    print("Hasta luego!") 

else:
    print("El valor que ingresó no es valido.")
```

5) Crear una lista con los nombres de 8 estudiantes presentes en clase.
• Preguntar al usuario si quiere agregar un nuevo estudiante o eliminar uno existente.
• Mostrar la lista final actualizada.

```python
# Ejercicio 5

lista = ["franco", "brisa", "francisco", "katerina", "judith", "mariel", "jorge", "fernando"]

print("Esta es la lista de los presentes:")

for nombre in lista:
    print(nombre)

print() # para dejar un espacio

respuesta = input("Si esea agregar algún nombre presione N\nSi desea eliminar uno existente presione E\n Si quiere salir presione S: ")
respuesta = respuesta.lower()

if respuesta == "n":
    nuevo_nom = input("Ingrese el nombre: ")
    lista.append(nuevo_nom)

elif respuesta == "e":
    sacar_nom = input("Ingrese el nombre que quiera sacar: ")
    lista.remove(sacar_nom)

elif respuesta == "s":
    print("Hasta luego!")

if respuesta == "n" or respuesta == "e":
    print("Lista actualizada:")
    for nombre in lista:
        print(nombre)
```

6) Dada una lista con 7 números, rotar todos los elementos una posición hacia la derecha (el
último pasa a ser el primero).

```python
# Ejercicio 6

lista = [5, 9, 7, 2, 4, 6, 99]
print(lista)

cant_elemen = len(lista)

for num in range(cant_elemen - 1):
    # lista[cant_elemen - 1 -  num] (para ir detectando donde va quedando el ultimo digito)
    # lista[cant_elemen - 2 -  num] (para detectar cual es el indice previo despues de cada iteracion)
    # luego se va haciendo el intercambio de posiciones, llevando el ultimo al principio y moviendo los demas a la derecha
    lista[cant_elemen - 1 -  num], lista[cant_elemen - 2 -  num] = lista[cant_elemen - 2 -  num], lista[cant_elemen - 1 -  num]

print("==== Nueva Lista ====")
print(lista)
```

semana.

8) Crear una matriz con las notas de 5 estudiantes en 3 materias.
• Mostrar el promedio de cada estudiante.
• Mostrar el promedio de cada materia.

```python
# Ejercicio 8

promedio = 0
matematica = 0
programacion = 0
ingles = 0
notas = [[7, 9, 10], [6, 7, 8], [4, 7, 6], [8, 6, 10], [5, 7, 8]]

print("En Matemática, Programación e inglés")
for fila in range(len(notas)):
    print(f"El Alumno {fila + 1} obtuvo:")

    for columna in range(len(notas[0])):
        print(notas[fila][columna], end=' ')

    print()
print("===============================")

print("En promedio")
for fila in range(len(notas)):
    print(f"El Alumno {fila + 1} obtuvo:")

    for columna in range(len(notas[0])):
        promedio += notas[fila][columna] # sumo las notas de cada estudiante

    matematica += notas[fila][0] # cada columna corresponde a la materia
    programacion += notas[fila][1]
    ingles += notas[fila][2]

    print(round(promedio / len(notas[0]), 2))
    promedio = 0 
print("===============================")

print("los promedios por materia son: ")
print(f"Matemática {matematica / (len(notas) + 1)}")
print(f"programación {programacion / (len(notas) + 1)}")
print(f"inglés {ingles / (len(notas) + 1)}")
```

10) Una tienda registra las ventas de 4 productos durante 7 días, en una matriz de 4x7.
• Mostrar el total vendido por cada producto.
• Mostrar el día con mayores ventas totales.
• Indicar cuál fue el producto más vendido en la semana.

```python
# Ejercicio 10

dia_mas_ventas = 0
producto_mas_vendido = 0

ventas = [[10, 20, 15, 26, 9, 20, 18], [5, 10, 9, 8, 11, 7, 12], [15, 10, 12, 14, 13, 11, 16], [10, 14, 9, 11, 10, 13, 17]]


for fila in range(len(ventas)):
    print(f"{fila + 1}º Producto")

    total_ventas = 0

    for columna in range(len(ventas[0])):
        print(ventas[fila][columna], end=' ')
        total_ventas += ventas[fila][columna]
    
    print(f"\nEl total vendido del articulo es: {total_ventas}")
    # registramos cual fue el producto mas vendido
    if total_ventas > producto_mas_vendido:
        producto_mas_vendido = total_ventas
        producto = fila +1
    
    print("====================")

# invierto el doble loop para sumar los datos de cada columna individualmente
for columna in range(len(ventas[0])):
    venta_dia = 0

    for fila in range(len(ventas)):
        venta_dia += ventas[fila][columna]

        if venta_dia > dia_mas_ventas:
            dia_mas_ventas = venta_dia
            num_dia = columna + 1

print(f"El producto mas vendido fue el {producto}º con {producto_mas_vendido} ventas.")
print(f"El día con mas ventas fue el {num_dia}º con {dia_mas_ventas} ventas.")
```
